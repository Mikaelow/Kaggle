```python
# uploading dataset
db_source = '/kaggle/input/cars-formated/cars_formated.csv'
# puting data base to the 
cars_formated = pd.read_csv(db_source, index_col = 'Unnamed: 0')
```

```python
cars_formated
```

```python
# choosing most value columns from the data base
columns_choosen = cars_formated.columns
columns_choosen = columns_choosen.drop('vin')
```

```python
cars_formated.dtypes
```

```python
one_hot_encoded_db = pd.get_dummies(cars_formated[columns_choosen])

one_hot_encoded_db
```

# Spliting data to features and labels

```python
features = one_hot_encoded_db.columns
# spliting to feature data base
# X = cars_formated[features]
features = features.drop('sellingprice')
features
```

```python
# spliting to label
X = one_hot_encoded_db[features]
y = one_hot_encoded_db['sellingprice']
X
```

# Building first model

```python
# importing models
from sklearn.tree import DecisionTreeRegressor
# import mae
from sklearn.metrics import mean_absolute_error
# import db split to validate data
from sklearn.model_selection import train_test_split
```

```python
# define model. Specify a number for random_state to ensure same results each run
car_model = DecisionTreeRegressor(random_state=1)

# fit model
car_model.fit(X, y)
```

```python
# predicting price
pred_car_price = car_model.predict(X)
# checking mae
mae_pred = mean_absolute_error(y,pred_car_price)
mae_pred
```

```python
# % of the mean
(mae_pred/y.mean())*100
```

**Observation: MAE is very low, only 0.309, it represents just 0.002% of the average car price.**

# Validating model

```python
# building new model

# spliting data to train and validation
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# define new car model
car_model_with_validation = DecisionTreeRegressor(random_state = 1)
# fiting model 
car_model_with_validation.fit(train_X, train_y)
# predicting price
pred_car_price_with_validation = car_model_with_validation.predict(val_X)
```

```python
# checking mae
mae_pred_with_validation = mean_absolute_error(val_y, pred_car_price_with_validation)
mae_pred_with_validation
```

```python
# % of the mean
(mae_pred_with_validation/y.mean())*100
```

**Observation: MAE is much bigger than without validation 2003.38, it represents just 14.66% of the average car price.**
