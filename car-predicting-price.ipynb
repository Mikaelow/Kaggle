```python
# uploading dataset
db_source = '/kaggle/input/cars-formated/cars_formated.csv'
# puting data base to the 
cars_formated = pd.read_csv(db_source, index_col = 'Unnamed: 0')
```

```python
cars_formated
```

```python
# choosing most value columns from the data base
columns_choosen = cars_formated.columns
columns_choosen = columns_choosen.drop('vin')
```

```python
cars_formated.dtypes
```

```python
# encoding from object to bool
one_hot_encoded_db = pd.get_dummies(cars_formated[columns_choosen])

one_hot_encoded_db
```

# Spliting data to features and labels

```python
features = one_hot_encoded_db.columns
# spliting to feature data base
# X = cars_formated[features]
features = features.drop('sellingprice')
features
```

```python
# spliting to label
X = one_hot_encoded_db[features]
y = one_hot_encoded_db['sellingprice']
X
```

# Building first model

```python
# importing models
from sklearn.tree import DecisionTreeRegressor
# import mae
from sklearn.metrics import mean_absolute_error
# import db split to validate data
from sklearn.model_selection import train_test_split
```

```python
# define model. Specify a number for random_state to ensure same results each run
car_model = DecisionTreeRegressor(random_state=1)

# fit model
car_model.fit(X, y)
```

```python
# predicting price
pred_car_price = car_model.predict(X)
# checking mae
mae_pred = mean_absolute_error(y,pred_car_price)
mae_pred
```

```python
# % of the mean
mae_pred_proc = (mae_pred/y.mean())*100
mae_pred_proc
```

**Observation: MAE is very low, only `0.309`, it represents just `0.002%` of the average car price.**

# Validating model

```python
# building new model

# spliting data to train and validation
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# define new car model
car_model_with_validation = DecisionTreeRegressor(random_state = 1)
# fiting model 
car_model_with_validation.fit(train_X, train_y)
# predicting price
pred_car_price_with_validation = car_model_with_validation.predict(val_X)
```

```python
# checking mae
mae_pred_with_validation = mean_absolute_error(val_y, pred_car_price_with_validation)
mae_pred_with_validation
```

```python
# % of the mean
mae_pred_with_validation_proc = (mae_pred_with_validation/y.mean())*100
mae_pred_with_validation_proc
```

**Observation: MAE is much bigger than without validation `2003.38`, it represents just `14.66%` of the average car price.**

# Underfitting and Overfitting

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
my_mae_proc = (my_mae  /y.mean())*100
my_mae_proc
```

**Observation: From the `max_leaf_nodes = [5, 50, 500, 5000]`, the best MAE occurs when `max_leaf_nodes` is equal to `5000`, with a value of `1750`, representing just `12.82%`.**

# Creating random forest

```python
# importing random forest
from sklearn.ensemble import RandomForestRegressor

# define random model
forest_model = RandomForestRegressor(random_state=1)
# fiting model
forest_model.fit(train_X, train_y)
# predicting 
car_price_forest_pred = forest_model.predict(val_X)
# compare car_price_forest_pred with val_y in mae
mae_random_forest = mean_absolute_error(val_y, car_price_forest_pred)
```

```python
mae_random_forest_proc = (mae_random_forest/y.mean())*100
mae_random_forest_proc
```

**Observation: RandomForestRegressor have the best from previous method MAE equal `1515`,representing just `11.1%`.**

# Create Linear Regression

```python
# importing Linear Regression
from sklearn.linear_model import LinearRegression

# creating Linear Regression model
linear_regression_model = LinearRegression()

# fiting model
linear_regression_model.fit(train_X, train_y)

# predicting price
pred_price_linear_regression = linear_regression_model.predict(val_X)

# compare car_price_forest_pred with val_y in mae
mae_linear_regression = mean_absolute_error(val_y, pred_price_linear_regression)
mae_linear_regression
```

```python
mae_linear_regression_proc = (mae_linear_regression/y.mean())*100
mae_linear_regression_proc
```
